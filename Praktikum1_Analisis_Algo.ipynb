```python
# y = number of element in array
def Sum(ab, y):
  sum = 0
  for i in range(y):
    sum = sum + ab[i]
  return sum

ab = [5, 6, 2, 1] # array / matriks
y = len(ab) # len = panjang data / matriks
print(Sum (ab, y ))
```

mencari jumlah semua elemen dari sebuah matriks

```python
x = 3
y = 3
ab = [(3, 2, 7),(2, 6, 8),(5, 1, 9)]
sum = 0

# melakukan iterasi data semua array 1-D dalam array 2-D
for i in range(x):

  # mencetak semua elemen dalam array 1-D ke i
  for j in range(y):

    # mencetak elemen ke-j dari baris ke-1
    sum += ab[i][j]

print(sum)
```

```python
def binary_search(sorted_list, target):
  low = 0
  high = len(sorted_list) - 1

  while low <= high:
    mid = (low + high) // 2 # temukan indeks tengah
    guess = sorted_list[mid]

    if guess == target:
      return mid # terget ditemukan, kembalikan indeks

    if guess > target:
      high = mid - 1 # target berada dibagian kiri

    else:
      low = mid + 1 #target berada dibagian kanan

  return - 1 # target tidak ada dalam daftar

# contoh input dan output
my_list = [1, 3, 5, 7, 9, 11, 13, 15]
target_val = 7

result = binary_search(my_list, target_val)

if result != -1:
  print(f"input: {my_list}, target: {target_val}")
  print(f"output: element ditemukan dalam index {result}")
else:
  print("output: element tidak ditemukan")
```

Fungsi untuk mencari bilangan faktorial

Big(o) nya adalah O(n!)

```python
y = int(input("Masukkan angka faktorialnya: "))

def faktorial(y):
  if y == 1:
    return 1
  else:
    return y * faktorial(y - 1)

print(faktorial(y))
```
