**BUBBLE SORT**

Algoritma paling sederhana yang bekerja dengan membandingkan dan menukar pasangan elemen berdekatan secara berulang hingga terurut.

Memiliki Kompleksitas **O(n^2)**

Prosedur BubbleSort(A : list of sortable items)

    n = panjang(A)
    Ulangi i dari 0 sampai n-1:
        Ulangi j dari 0 sampai n-i-2:
            Jika A[j] > A[j+1] maka:
                Tukar(A[j], A[j+1])

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
```

**SELECTION SORT**

Mencari elemen terkecil/terbesar dari bagian data yang belum terurut dan menukarnya ke posisi yang tepat secara bertahap.

Memiliki Kompleksitas **O(n^2)**

Prosedur SelectionSort(A : list of sortable items)

    n = panjang(A)
    Ulangi i dari 0 sampai n-1:
        indeks_min = i
        Ulangi j dari i+1 sampai n-1:
            Jika A[j] < A[indeks_min] maka:
                indeks_min = j
        Tukar(A[i], A[indeks_min])

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
```

**INSERTION SORT**

Membangun array terurut satu per satu dengan mengambil elemen dan menyisipkannya ke posisi yang tepat di antara elemen yang sudah terurut.

Memiliki Kompleksitas **O(n^2)**

Prosedur InsertionSort(A : list of sortable items)

    Ulangi i dari 1 sampai panjang(A)-1:
        key = A[i]
        j = i - 1
        Selama j >= 0 dan A[j] > key:
            A[j + 1] = A[j]
            j = j - 1
        A[j + 1] = key

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
```

**QUICK SORT**

 Metode cepat yang memilih elemen sebagai pivot dan mempartisi array berdasarkan nilai pivot (lebih kecil di kiri, lebih besar di kanan).

 Memiliki Kompleksitas **O(n \log n)**

Prosedur QuickSort(A, rendah, tinggi)

    Jika rendah < tinggi:
        p = Partisi(A, rendah, tinggi)
        QuickSort(A, rendah, p - 1)
        QuickSort(A, p + 1, tinggi)

Fungsi Partisi(A, rendah, tinggi):

    pivot = A[tinggi]
    i = rendah - 1
    Ulangi j dari rendah sampai tinggi - 1:
        Jika A[j] < pivot:
            i = i + 1
            Tukar(A[i], A[j])
    Tukar(A[i + 1], A[tinggi])
    Kembalikan i + 1

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[len(arr) // 2]
        left = [x for x in arr if x < pivot]
        middle = [x for x in arr if x == pivot]
        right = [x for x in arr if x > pivot]
        return quick_sort(left) + middle + quick_sort(right)
```
