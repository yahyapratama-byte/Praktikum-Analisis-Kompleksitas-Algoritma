```python
# Recursive Binary String
def binary_string(n, hasil=""):

  if n == 0:
    print(hasil)
    return

  binary_string(n-1, hasil + "0")
  binary_string(n-1, hasil + "1")

binary_string(3)
```

```python
# Brute Force Password
import itertools
import string

chars = "abc"

for panjang in range(1, 4):

  for kombinasi in itertools.product(chars, repeat=panjang):

    password = ''.join(kombinasi)
    print(password)
```

```python
# Tower of Hanoi
def hanoi(n, asal, bantu, tujuan):

  if n == 1:
    print("Pindah disk 1 dari", asal, "ke", tujuan)
    return

  hanoi(n-1, asal, tujuan, bantu)

  print("Pindah disk", n, "dari", asal, "ke", tujuan)

  hanoi(n-1, bantu, asal, tujuan)

hanoi(3, 'A', 'B', 'C')
```

TUGAS : CARILAH ALGORITMA dengan rumus euclidean distance

```python
import math

def euclidean_distance(titik1, titik2):

    if len(titik1) != len(titik2):
        raise ValueError("Dimensi kedua titik harus sama!")

    total_kuadrat = 0

    for i in range(len(titik1)):
        total_kuadrat += (titik1[i] - titik2[i]) ** 2


    return math.sqrt(total_kuadrat)


titik_A = [1, 2, 3]
titik_B = [4, 5, 6]

jarak = euclidean_distance(titik_A, titik_B)
print(f"Jarak Euclidean: {jarak}")
```
