Mencari nilai maksimum

```python
#mencari nilai maksimum(megembalikan indeks)
def find_max(y):
  n = len(y)
  for i in range(n -1, -1, -1):   #dari n-1 ke 0
    if y[i] is not None:
      return i
  return None #jika list kosong atau semua none

#mengembalikan nilai elemen
def find_maximum_value(y):
  n = len(y)
  for i in range(n - 1, -1, -1):
    if y[i] is not None:
      return y[i]
  return None

x = [None, 5, None, 9, None, None]

print(find_max(x))   #output: 3 (indeks dari nilai 9)
print(find_maximum_value(x)) #output: 9 (nilai elemen itu sendiri)
```

Mencari Pseudecode dan Coding algoritma red-black tree kemudian analisa worth case dan average case nya

**Pseudecode pada red-black tree**

Algoritma RB-Insert(T, z):
  1. Lakukan BST Insert standar dan beri warna z = MERAH.
  2. SELAMA z bukan root DAN z.parent berwarna MERAH:
  
       a. JIKA z.parent adalah anak kiri dari z.grandparent:
          
          i.   y = paman z (anak kanan z.grandparent)
          
          ii.  JIKA y berwarna MERAH (Case 1):
                 - Ubah warna z.parent dan y menjadi HITAM
                 - Ubah warna z.grandparent menjadi MERAH
                 - z = z.grandparent
          iii. LAINNYA (Paman HITAM):
                 - JIKA z adalah anak kanan (Case 2):
                   - z = z.parent
                   - LEFT-ROTATE(T, z)
                 - (Case 3) Ubah warna z.parent jadi HITAM, z.grandparent jadi MERAH
                 - RIGHT-ROTATE(T, z.grandparent)
        b. LAINNYA (z.parent adalah anak kanan):
        
        (Lakukan simetri dari langkah a)

  3. Ubah warna ROOT menjadi HITAM

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.parent = None
        self.left = None
        self.right = None
        self.color = 1  # 1 untuk Merah, 0 untuk Hitam

class RedBlackTree:
    def __init__(self):
        self.TNULL = Node(0)
        self.TNULL.color = 0
        self.TNULL.left = None
        self.TNULL.right = None
        self.root = self.TNULL

    # Rotasi Kiri
    def left_rotate(self, x):
        y = x.right
        x.right = y.left
        if y.left != self.TNULL:
            y.left.parent = x

        y.parent = x.parent
        if x.parent == None:
            self.root = y
        elif x == x.parent.left:
            x.parent.left = y
        else:
            x.parent.right = y
        y.left = x
        x.parent = y

    # Rotasi Kanan
    def right_rotate(self, x):
        y = x.left
        x.left = y.right
        if y.right != self.TNULL:
            y.right.parent = x

        y.parent = x.parent
        if x.parent == None:
            self.root = y
        elif x == x.parent.right:
            x.parent.right = y
        else:
            x.parent.left = y
        y.right = x
        x.parent = y

    # Memperbaiki keseimbangan setelah insert
    def fix_insert(self, k):
        while k.parent.color == 1:
            if k.parent == k.parent.parent.right:
                u = k.parent.parent.left  # paman
                if u.color == 1:
                    u.color = 0
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    k.set = k.parent.parent
                else:
                    if k == k.parent.left:
                        k = k.parent
                        self.right_rotate(k)
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    self.left_rotate(k.parent.parent)
            else:
                u = k.parent.parent.right  # paman
                if u.color == 1:
                    u.color = 0
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    k = k.parent.parent
                else:
                    if k == k.parent.right:
                        k = k.parent
                        self.left_rotate(k)
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    self.right_rotate(k.parent.parent)
            if k == self.root:
                break
        self.root.color = 0

    # Fungsi Insert Utama
    def insert(self, key):
        node = Node(key)
        node.parent = None
        node.data = key
        node.left = self.TNULL
        node.right = self.TNULL
        node.color = 1

        y = None
        x = self.root

        while x != self.TNULL:
            y = x
            if node.data < x.data:
                x = x.left
            else:
                x = x.right

        node.parent = y
        if y == None:
            self.root = node
        elif node.data < y.data:
            y.left = node
        else:
            y.right = node

        if node.parent == None:
            node.color = 0
            return

        if node.parent.parent == None:
            return

        self.fix_insert(node)

    def print_tree(self, node, indent, last):
        if node != self.TNULL:
            print(indent, end="")
            if last:
                print("R----", end="")
                indent += "     "
            else:
                print("L----", end="")
                indent += "|    "

            s_color = "RED" if node.color == 1 else "BLACK"
            print(str(node.data) + "(" + s_color + ")")
            self.print_tree(node.left, indent, False)
            self.print_tree(node.right, indent, True)

# Contoh Penggunaan
if __name__ == "__main__":
    bst = RedBlackTree()
    nums = [10, 20, 30, 15, 25]
    for n in nums:
        bst.insert(n)

    print("Struktur Red-Black Tree:")
    bst.print_tree(bst.root, "", True)
```

Menganalisa worth case dan average case dari algoritma red-black tree pada pseudecode dan codingan diatas:

1.Average Case : O(log n)

  Pada average case elemen dimasukkan dalam urutan acak,  red-black tree akan membentuk struktur yang seimbang secara otomatis. karena tinggi pohon (h) proporsional dengan Log *n*, maka setiap operasi pencarian dan penyisipan hanya membutuhkan waktu sesuai dengan tinggi pohonnya.

2.Worth case: O(log n)

berikut keunggulan dalam algoritma red-black tree:

  a. Pada BST biasa, jika data dimasukkan secara berurutan    
  (misal: 1, 2, 3, 4, 5), pohon akan menjadi "miring" (skewed)   dan berubah menjadi seperti Linked List dengan kompleksitas  O(n).

  b.  Pada Red-Black Tree, aturan "Black Height" dan larangan dua node merah berurutan menjamin bahwa jalur terpanjang (dari root ke daun) tidak akan pernah lebih dari dua kali lipat panjang jalur terpendek.

  c. Secara matematis, tinggi maksimal Red-Black Tree adalah 2 \log_2(n + 1). Karena angka 2 adalah konstanta, maka dalam notasi Big O tetap ditulis sebagai O(log n).
