```python
import math

def euclidean_distance(a, b):
  jumlah = 0
  for i in range(len(a)):
    jumlah += (a[i] - b[i]) ** 2
  return math.sqrt(jumlah)

x = [1, 2]
y = [4, 6]
print(euclidean_distance(x, y))
```

```python
def manhattan_distance(a, b):
  total = 0
  for i in range(len(a)):
    total += abs(a[i] - b[i])
  return total

print(manhattan_distance([1, 2], [4,6]))
```

```python
import math

def cosine_similarity(a, b):
    dot = sum(a[i] * b[i] for i in range(len(a)))
    norm_a = math.sqrt(sum(x*x for x in a))
    norm_b = math.sqrt(sum(x*x for x in b))
    return dot / (norm_a * norm_b)

print(cosine_similarity([1, 2], [4, 6]))
```

```python
def hamming_distance(a, b):

  return sum(1 for i in range(len(a)) if a[i] != b[i])

print(hamming_distance("1011", "1001"))
```

```python
def permutasi(arr, l, r):
  if l == r:
    print(arr)
  else:
    for i in range(l, r + 1):
      #swap
      arr[l], arr[i] = arr[i], arr[l]

      #rekursi
      permutasi(arr, l + 1, r)

      #backtrack (kembalikan seperti semula)
      arr[l], arr[i] = arr[i], arr[l]

# contoh penggunaan
data = [1, 2, 3]
permutasi(data, 0, len(data) - 1)
```

Mencari jalur terpendek (shortest path) adalah masalah klasik dalam ilmu komputer. Algoritma yang paling populer dan efisien untuk graf dengan bobot non-negatif adalah Algoritma Dijkstra.

Berikut adalah rincian lengkapnya, mulai dari logika dasar hingga implementasi kode.

1. Pseudocode (Algoritma Dijkstra)
Algoritma ini bekerja dengan prinsip greedy, yaitu selalu memilih simpul dengan jarak terkecil yang belum dikunjungi.

Plaintext
PROSEDUR Dijkstra(Graf, Sumber):
    BUAT himpunan antrean_prioritas Q
    
    UNTUK SETIAP simpul v dalam Graf:
        jarak[v] = TAK_TERHINGGA
        sebelumnya[v] = UNDEFINED
        TAMBAHKAN v ke dalam Q
    
    jarak[Sumber] = 0
    
    SELAMA Q tidak kosong:
        u = simpul dalam Q dengan jarak[u] terkecil
        HAPUS u dari Q
        
        UNTUK SETIAP tetangga v dari u:
            alternatif = jarak[u] + bobot(u, v)
            JIKA alternatif < jarak[v]:
                jarak[v] = alternatif
                sebelumnya[v] = u
                PERBARUI v dalam Q dengan prioritas baru (alternatif)
                
    KEMBALIKAN jarak, sebelumnya

```python
import heapq

def dijkstra(graph, start):
    # Inisialisasi jarak dengan tak terhingga
    distances = {node: float('infinity') for node in graph}
    distances[start] = 0

    # Priority queue: (jarak, simpul)
    priority_queue = [(0, start)]

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        # Jika jarak di queue lebih besar dari jarak yang sudah ditemukan, abaikan
        if current_distance > distances[current_node]:
            continue

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            # Jika ditemukan jalur yang lebih pendek
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances

# Contoh Penggunaan:
graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'C': 5, 'D': 10},
    'C': {'E': 3},
    'D': {'F': 11},
    'E': {'D': 4},
    'F': {}
}

print(dijkstra(graph, 'A'))
```

Kompleksitas waktu Algoritma Dijkstra sangat bergantung pada struktur data yang digunakan untuk mengelola antrean prioritas.Jika menggunakan Min-Priority Queue (Binary Heap):Rumus kompleksitasnya adalah:$$O((V + E) \log V)$$
