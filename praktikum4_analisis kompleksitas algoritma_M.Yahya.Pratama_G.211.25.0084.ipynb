TENTUKAN KOMPLEKSITAS WAKTU BIG O DAN ANALISA CASE ALGORITMA PENCARIAN

```python
#Linear Search
def linear_search(arr, target):
  for i in range(len(arr)):
    if arr[i] == target:
      return i
  return -1
# contoh
data = [10, 20, 30, 40, 50]
print(linear_search(data, 30)) # output: 2
```

**Linear Search**

kompleksitas waktunya (Big O) dalam Linear Search adalah O(n)

best case: O(1) apabila yang dicari (target) berada di index pertama

worst case: O(n) apabila target yang dicari berada di posisi terakhir atau tidak ada sama sekali di dalam array (harus diperiksa satu persatu sampai tuntas)

average case: O(n) pencarian akan terhenti di tengah-tengah, tetapi dalam notasi Big O. Faktor konstan diabaikan

```python
#Binary search (Iteratif)
def binary_search(arr, target):
  left, right = 0, len(arr) - 1

  while left <= right:
    mid = (left + right) // 2

    if arr[mid] == target:
      return mid
    elif arr[mid] < target:
      left = mid + 1
    else:
      right = mid - 1

  return - 1

#contoh (data harus terurut)
data = [10, 20, 30, 40, 50]
print(binary_search(data, 40)) # output: 3
```

```python
#Binary Search (rekursif)
def binary_search_recursive(arr, left, right, target):
  if left > right:
    return -1

  mid = (left + right) // 2

  if arr[mid] == target:
    return mid
  elif arr[mid] < target:
    return binary_search_recursive(arr, mid + 1, right, target)
  else:
    return binary_search_recursive(arr, left, mid - 1, target)

#contoh
data = [10, 20, 30, 40, 50]
print(binary_search_recursive(data, 0, len(data) - 1, 20)) #output: 1
```

**Binary Search (Iteratif & Rekursif)**

Algoritma ini bekerja dengan membagi ruang pencarian menjadi dua pada setiap langkah.

Worst Case ($O(\log n)$): Terjadi jika target berada di ujung pembelahan terakhir atau tidak ditemukan. Karena data terus dibagi dua, jumlah langkah berkurang secara logaritmik.

Best Case ($O(1)$): Terjadi jika nilai tengah (mid) langsung sesuai dengan target pada percobaan pertama.

Average Case ($O(\log n)$): Rata-rata pencarian dalam struktur pohon biner seimbang.

```python
# Jump Search

import math

def jump_search(arr, target):
  n = len(arr)
  step = int(math.sqrt(n))
  prev = 0

  # lompat blok
  while arr[min(step, n) - 1] < target:
    prev = step
    step += int(math.sqrt(n))
    if prev >= n:
      return -1

    # linear search di blok
    for i in range(prev, min(step, n)):
      if arr[i] == target:
       return i

  return -1

# contoh
data = [10, 20, 30, 40, 50, 60, 70]
print(jump_search(data, 60)) # output: 5

```

**Jump Search**

Algoritma ini melompat dalam blok-blok berukuran $\sqrt{n}$ sebelum melakukan linear search di dalam blok yang relevan.

Worst Case ($O(\sqrt{n})$): Terjadi jika target berada di blok terakhir atau elemen terbesar. Kita melakukan $n/m$ lompatan dan $m-1$ pencarian linier di dalam blok (di mana $m = \sqrt{n}$).

Best Case ($O(1)$): Terjadi jika elemen pertama langsung merupakan target.

Average Case ($O(\sqrt{n})$): Efisiensinya berada di antara Linear Search dan Binary Search.

```python
# Interpolation Search

def interpolation_search(arr, target):
  low = 0
  high = len(arr) - 1

  while low <= high and target >= arr[low] and target <= arr[high]:
    if low == high:
      if arr[low] == target:
        return low
      return - 1

    #rumus interpolasi
    pos = low + ((target - arr[low]) * (high - low) // (arr[high] - arr[low]))

    if arr[pos] == target:
      return pos
    elif arr[pos] < target:
      low = pos + 1
    else:
      high = pos - 1

  return - 1

#contoh
data = [10, 20, 30, 40, 50]
print(interpolation_search(data, 50)) # output: 4
```

**Interpolation Search**

Ini adalah pengembangan dari Binary Search yang memprediksi posisi target menggunakan rumus estimasi (mirip seperti mencari kata di kamus).

Worst Case ($O(n)$): Terjadi jika distribusi data tidak merata (tidak uniform), sehingga rumus interpolasi terus meleset dan hanya mengurangi satu elemen setiap langkah.

Best Case ($O(1)$): Terjadi jika target ditemukan langsung pada estimasi posisi pertama.

Average Case ($O(\log(\log n))$): Sangat cepat jika data terdistribusi secara merata (uniform distribution).
