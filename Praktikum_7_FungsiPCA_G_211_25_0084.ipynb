```python
# Algoritma fungsi PCA
def absolut(x):
    return x if x >= 0 else -x

def akar_kuadrat(x, tol=1e-12):

    if x == 0:
        return 0.0
    tebakan = x / 2.0
    while True:
        berikutnya = 0.5 * (tebakan + x / tebakan)
        if absolut(berikutnya - tebakan) < tol:
            return berikutnya
        tebakan = berikutnya

def hitung_rata_rata(matriks):

    n_baris = len(matriks)
    n_kolom = len(matriks[0])
    rata_rata = [0.0] * n_kolom
    for j in range(n_kolom):
        total = sum(matriks[i][j] for i in range(n_baris))
        rata_rata[j] = total / n_baris
    return rata_rata

def center_data(matriks, rata_rata):
    return [[matriks[i][j] - rata_rata[j] for j in range(len(rata_rata))] for i in range(len(matriks))]

def transpose(matriks):
    return [[matriks[i][j] for i in range(len(matriks))] for j in range(len(matriks[0]))]

def perkalian_matriks(A, B):

    m, n, p = len(A), len(A[0]), len(B[0])
    hasil = [[0.0] * p for _ in range(m)]
    for i in range(m):
        for j in range(p):
            hasil[i][j] = sum(A[i][k] * B[k][j] for k in range(n))
    return hasil

def dot_product(v1, v2):
    return sum(x * y for x, y in zip(v1, v2))

def norma_vektor(v):
    return akar_kuadrat(sum(x**2 for x in v))

def normalisasi_vektor(v):

    norm = norma_vektor(v)
    if norm == 0:
        return v
    return [x / norm for x in v]

def power_iteration(A, max_iter=1000, tol=1e-9):

    n = len(A)
    v = [1.0] * n
    v = normalisasi_vektor(v)

    eigenvalue_lama = 0.0

    for _ in range(max_iter):

        Av = [sum(A[i][j] * v[j] for j in range(n)) for i in range(n)]

        eigenvalue = dot_product(v, Av)
        v = normalisasi_vektor(Av)
        if absolut(eigenvalue - eigenvalue_lama) < tol:
            break
        eigenvalue_lama = eigenvalue

    return eigenvalue, v

def hitung_eigen(matriks_kovarian, n_komponen):

    A = [baris[:] for baris in matriks_kovarian]
    n = len(A)
    eigenvalues = []
    eigenvectors = []

    for _ in range(n_komponen):
        val, vec = power_iteration(A)
        eigenvalues.append(val)
        eigenvectors.append(vec)

        for i in range(n):
            for j in range(n):
                A[i][j] -= val * vec[i] * vec[j]

    return eigenvalues, eigenvectors

def PCA_total_manual(X, n_komponen):
    n_sampel = len(X)
    n_fitur = len(X[0])

    rata_rata = hitung_rata_rata(X)
    X_centered = center_data(X, rata_rata)

    X_T = transpose(X_centered)
    matriks_kovarian = perkalian_matriks(X_T, X_centered)

    for i in range(n_fitur):
        for j in range(n_fitur):
            matriks_kovarian[i][j] /= (n_sampel - 1)

    eigenvalues, eigenvectors = hitung_eigen(matriks_kovarian, n_komponen)

    W = transpose(eigenvectors)
    X_pca = perkalian_matriks(X_centered, W)

    return X_pca, eigenvalues, eigenvectors

dataset = [
    [170, 65, 25],
    [160, 55, 30],
    [180, 80, 22],
    [150, 50, 45],
    [175, 70, 28]
]

n_target = 2
hasil_proyeksi, nilai_eigen, vektor_eigen = PCA_total_manual(dataset, n_target)

print("FUNGSI PCA")
print("\nEigenvalues (Besar Varians):")
for i, val in enumerate(nilai_eigen):
    print(f"Komponen Utama {i+1}: {val:.4f}")

print("\nData Setelah Reduksi (2 Dimensi):")
for baris in hasil_proyeksi:

    baris_terformat = [float(f"{elemen:.4f}") for elemen in baris]
    print(baris_terformat)
```
